```python
import numpy as np#Digunakan untuk operasi numerik, termasuk membuat array dan fungsi matematika seperti sin.

def integrate_rec(f, a, b, n): #Mendefinisikan fungsi bernama integrate_rec. Fungsi ini menerima empat argumen f: Fungsi yang ingin diintegralkan (misalnya sin).
                                                                                                               #a: Batas bawah interval.
                                                                                                               #b: Batas atas interval.
                                                                                                               #n: Jumlah persegi panjang (sub-interval) yang digunakan.
    h = (b-a)/n #Menghitung lebar setiap persegi panjang ($h$ atau $\Delta x$).
    x = np.linspace(a, b, n+1) #Membuat titik-titik koordinat pada sumbu x dari $a$ sampai $b$. Karena ada $n$ bidang, maka dibutuhkan $n+1$ titik potong.
    i=0 #Inisialisasi variabel penghitung (counter) untuk perulangan.
    area=0 #Inisialisasi variabel untuk menyimpan total luas yang akan dijumlahkan.
    while i<n: #Melakukan perulangan sebanyak $n$ kali (untuk setiap persegi panjang).
        sup_rect =  f(x[i])*h #Menghitung luas satu persegi panjang. Rumusnya adalah $Tinggi \times Lebar$.
        area += sup_rect #Menambahkan luas persegi panjang yang baru dihitung ke dalam total area.
        i += 1 #Menambah nilai counter agar perulangan berlanjut ke titik berikutnya.
    return area# Mengembalikan hasil akhir total luas setelah perulangan selesai.
'''
We test the Rectangle method on the sine funcion were the definite integral in the interval [0, %*\textcolor{codeviolet}{$\pi$}*)/2] is equal to 1. 
'''

sup_5 = integrate_rec(np.sin, 0, np.pi/2, 5) #Memanggil fungsi integrate_rec dengan jumlah persegi panjang yang berbeda (5, 10, dan 100).
sup_10 = integrate_rec(np.sin, 0, np.pi/2, 10) 
sup_100 = integrate_rec(np.sin, 0, np.pi/2, 100) 

print('Using n=5, the rectangle method returns a value of {:.2f}'.format(sup_5)) #Menampilkan hasil perhitungan ke layar dengan format dua angka di belakang koma (:.2f).
print('Using n=10, the rectangle method returns a value of {:.2f}'.format(sup_10)) #Menampilkan hasil perhitungan untuk n=10.
print('Using n=100, the rectangle method returns a value of {:.2f}'.format(sup_100))
     
'''
Output:
Using n=5, the rectangle method returns a value of 0.83
Using n=10, the rectangle method returns a value of 0.92
Using n=100, the rectangle method returns a value of 0.99
'''
 
```

```python
import numpy as np#Digunakan untuk operasi numerik, termasuk membuat array dan fungsi matematika seperti sin.

def integrate_trap(f, a, b, n): #Mendefinisikan fungsi bernama integrate_trap. Fungsi ini menerima empat argumen f: Fungsi yang ingin diintegralkan (misalnya sin).
    # Implementation of the trapezoidal rule
    h = (b-a)/n #Menghitung lebar setiap sub-interval (atau persegi panjang) dengan rumus $h = \frac{b-a}{n}$, di mana $b$ adalah batas atas, $a$ adalah batas bawah, dan $n$ adalah jumlah sub-interval.
    x = np.linspace(a, b, n+1) #Membuat array x yang berisi titik-titik pada interval [a, b] dengan jumlah total n+1 titik. Fungsi np.linspace digunakan untuk menghasilkan titik-titik ini secara merata.
    i=1 #Inisialisasi variabel penghitung (counter) untuk perulangan, dimulai dari 1 karena titik pertama (x[0]) sudah digunakan dalam perhitungan awal.
    area = h*(f(x[0]) + f(x[n]))/2 #Menghitung luas awal menggunakan rumus trapezoidal untuk titik pertama dan terakhir. Rumusnya adalah $Area = \frac{h}{2} [f(x_0) + f(x_n)]$, di mana $x_0$ adalah titik pertama dan $x_n$ adalah titik terakhir.
    while i<n: #Melakukan perulangan sebanyak n-1 kali (karena titik pertama dan terakhir sudah dihitung).
        sup_rect = f(x[i])*h #Menghitung luas untuk setiap titik tengah (x[i]) menggunakan rumus $Area = f(x_i) \cdot h$, di mana $f(x_i)$ adalah nilai fungsi pada titik x[i] dan $h$ adalah lebar sub-interval.
        area += sup_rect  #Menambahkan luas yang dihitung untuk titik tengah ke total area. Ini dilakukan untuk setiap titik tengah dalam interval.
        i += 1  #Menambah nilai counter agar perulangan berlanjut ke titik berikutnya.
    return area # Mengembalikan hasil akhir total luas setelah perulangan selesai.

'''
We test the trapezoidal rule on the known sine funcion were the 
definite integral in the interval [0, %*\textcolor{codeviolet}{$\pi$}*)/2] is equal to 1. 
'''

sup_5 = integrate_trap(np.sin, 0, np.pi/2, 5) #Memanggil fungsi integrate_trap dengan jumlah sub-interval yang berbeda (5 dan 10).
sup_10 = integrate_trap(np.sin, 0, np.pi/2, 10) #Memanggil fungsi integrate_trap untuk n=10.

print('Using n=5, the trapezoidal rule returns a value of {:.2f}'.format(sup_5)) #Menampilkan hasil perhitungan ke layar dengan format dua angka di belakang koma (:.2f).
print('Using n=10, the trapezoidal rule returns a value of {:.2f}'.format(sup_10)) #Menampilkan hasil perhitungan untuk n=10.
   
'''
Output:
Using n=5, the trapezoidal rule returns a value of 0.99
Using n=10, the trapezoidal rule returns a value of 1.00
'''
 
```

```python
import numpy as np #Digunakan untuk operasi numerik, termasuk membuat array dan fungsi matematika seperti sin.
from scipy import integrate #Digunakan untuk melakukan integrasi numerik, termasuk metode trapezoidal dan Simpson.


x = np.linspace(0,9, 3) # 3 divisions %*\textcolor{codegreen}{$[x_0,x_1,x_2]$}*), n=2  #Membuat array x yang berisi 3 titik yang terdistribusi secara merata antara 0 dan 9. Fungsi np.linspace digunakan untuk menghasilkan titik-titik ini, dan jumlah titik yang dihasilkan adalah n+1 (dalam hal ini, 3 titik untuk n=2).
y = x**2 #Membuat array y yang berisi nilai fungsi x^2 untuk setiap elemen dalam array x. Ini berarti bahwa y[i] akan menjadi kuadrat dari x[i], sehingga menghasilkan nilai fungsi yang sesuai untuk setiap titik dalam array x.

sup_trapz = integrate.trapezoid(y,x) #Menghitung integral numerik menggunakan metode trapezoidal dengan fungsi integrate.trapezoid dari modul scipy.integrate. Fungsi ini mengambil dua argumen: y, yang berisi nilai fungsi pada titik-titik x, dan x, yang berisi titik-titik di mana fungsi dievaluasi. Hasilnya disimpan dalam variabel sup_trapz.
sup_simps = integrate.simpson(y,x) #Menghitung integral numerik menggunakan metode Simpson dengan fungsi integrate.simpson dari modul scipy.integrate. Fungsi ini juga mengambil dua argumen: y, yang berisi nilai fungsi pada titik-titik x, dan x, yang berisi titik-titik di mana fungsi dievaluasi. Hasilnya disimpan dalam variabel sup_simps.


print('Using n=2, the trapezoidal rule returns a value of {:.0f}'.format(sup_trapz)) #Menampilkan hasil perhitungan metode trapezoidal ke layar dengan format angka bulat (:.0f).
print('Using n=2, the composite Simpson rule returns a value of {:.0f}'.format(sup_simps)) #Menampilkan hasil perhitungan metode Simpson ke layar dengan format angka bulat (:.0f).

'''
Output:
Using n=2, the trapezoidal rule returns a value of 273
Using n=2, the composite Simpson rule returns a value of 243
'''
```

```python
def pressure(r, ro, g): #Mendefinisikan fungsi bernama pressure yang akan menghitung tekanan litostatik (p) pada berbagai jarak dari pusat bumi (r) berdasarkan kepadatan (ro) dan percepatan gravitasi (g). Fungsi ini menerima tiga argumen: r, yang merupakan array jarak dari pusat bumi dalam kilometer; ro, yang merupakan array kepadatan untuk setiap lapisan interior bumi; dan g, yang merupakan array percepatan gravitasi untuk setiap jarak dari pusat bumi. Fungsi ini akan menggunakan metode trapezoidal untuk menghitung integral numerik yang diperlukan untuk menentukan tekanan litostatik pada setiap jarak.
    
    p = np.zeros(len(r)) #Membuat array p yang berisi nilai-nilai nol dengan panjang yang sama dengan array r. Array ini akan digunakan untuk menyimpan nilai tekanan litostatik yang dihitung untuk setiap jarak dalam array r.
    r = r *1000 # from Km to m Mengubah nilai-nilai dalam array r dari kilometer ke meter dengan mengalikan setiap elemen dalam array r dengan 1000. Ini penting karena perhitungan tekanan litostatik biasanya dilakukan dalam satuan meter, sehingga kita perlu memastikan bahwa jarak yang digunakan dalam perhitungan sesuai dengan satuan yang diharapkan.
    
    for i in range(0,len(r)): #Melakukan perulangan untuk menghitung tekanan litostatik pada setiap jarak dalam array r, dimulai dari indeks 0 hingga panjang array r. Perulangan ini akan memungkinkan kita untuk menghitung nilai p untuk setiap jarak yang diberikan dalam array r. 
        r1 = r[i:len(r)] #Membuat array r1 yang berisi nilai-nilai dari array r mulai dari indeks i hingga indeks terakhir (len(r)-1). Ini berarti bahwa r1 akan berisi semua jarak dari pusat bumi mulai dari jarak saat ini yang sedang dihitung dalam perulangan hingga jarak terakhir dalam array r.
        ro1 = ro[i:len(r)]  #Membuat array ro1 yang berisi nilai-nilai kepadatan (ro) yang sesuai dengan jarak dalam array r1. Ini berarti bahwa ro1 akan berisi nilai kepadatan untuk semua lapisan interior bumi yang berada pada jarak dari pusat bumi mulai dari jarak saat ini yang sedang dihitung dalam perulangan hingga jarak terakhir dalam array r.
        g1 = g[i:len(r)] #Membuat array g1 yang berisi nilai-nilai percepatan gravitasi (g) yang sesuai dengan jarak dalam array r1. Ini berarti bahwa g1 akan berisi nilai percepatan gravitasi untuk semua jarak dari pusat bumi mulai dari jarak saat ini yang sedang dihitung dalam perulangan hingga jarak terakhir dalam array r.
        y = ro1*g1 #Menghitung array y yang merupakan hasil perkalian antara array ro1 (kepadatan) dan array g1 (percepatan gravitasi). Ini mungkin mewakili kontribusi tekanan litostatik dari setiap lapisan interior bumi pada jarak tertentu.
        p1 = trapezoid(y,r1) #Menghitung integral numerik dari array y terhadap array r1 menggunakan metode trapezoidal dengan fungsi trapezoid dari modul scipy.integrate. Hasilnya disimpan dalam variabel p1, yang mungkin mewakili total tekanan litostatik pada jarak saat ini (r[i]) berdasarkan kontribusi dari semua lapisan interior bumi hingga jarak terakhir dalam array r.
        p[i] = p1 #Menetapkan nilai p1 yang telah dihitung ke dalam array p pada indeks i. Ini berarti bahwa untuk setiap jarak dalam array r, kita akan menyimpan nilai tekanan litostatik yang sesuai dalam array p, sehingga pada akhir perulangan, array p akan berisi nilai tekanan litostatik untuk setiap jarak dari pusat bumi yang diberikan dalam array r.
    return p # Mengembalikan array p yang berisi nilai tekanan litostatik untuk setiap jarak dari pusat bumi yang diberikan dalam array r. Array ini dapat digunakan untuk analisis lebih lanjut atau visualisasi.
    
p = pressure(r,ro,g)/1e9 # expressed in GPa Memanggil fungsi pressure dengan argumen r, ro, dan g untuk menghitung tekanan litostatik (p) pada berbagai jarak dari pusat bumi. Hasilnya disimpan dalam variabel p, yang kemudian dibagi dengan 1e9 untuk mengubah satuan dari pascal (Pa) ke gigapascal (GPa), sehingga nilai yang disimpan dalam p akan lebih mudah dibaca dan diinterpretasikan dalam konteks geologi atau fisika bumi.
z = np.linspace(6400, 1, 6400) #Membuat array z yang berisi nilai-nilai dari 6400 hingga 1 dengan total 6400 elemen. Fungsi np.linspace digunakan untuk menghasilkan nilai-nilai ini secara merata antara 6400 dan 1, yang mungkin mewakili kedalaman dari permukaan bumi ke pusat bumi dalam kilometer. Array z ini akan digunakan untuk menggambarkan hubungan antara kedalaman (z) dan tekanan litostatik (p) dalam grafik selanjutnya.

fig, ax = plt.subplots() #Membuat sebuah figure dan axes menggunakan fungsi plt.subplots() dari modul matplotlib.pyplot. Fungsi ini mengembalikan dua objek: fig, yang merupakan figure utama untuk grafik, dan ax, yang merupakan axes (sumbu) tempat kita akan menggambar grafik. Dengan menggunakan ax, kita dapat menambahkan elemen-elemen seperti garis, label, dan grid ke dalam grafik yang akan dibuat.
ax.plot(z,p) #Menggambar grafik garis menggunakan metode plot dari objek ax. Grafik ini akan menampilkan hubungan antara kedalaman dari permukaan bumi (z) dan tekanan litostatik (p) pada berbagai lapisan interior bumi. Tidak ada label yang diberikan untuk grafik ini, sehingga tidak akan ada informasi tambahan yang ditampilkan di legenda.
ax.grid() #Menambahkan grid ke grafik menggunakan metode grid dari objek ax. Grid ini akan membantu dalam membaca nilai-nilai pada grafik dengan lebih mudah, karena garis-garis grid akan memudahkan untuk melihat di mana titik-titik data berada dalam kaitannya dengan sumbu x dan y. Grid biasanya terdiri dari garis horizontal dan vertikal yang membentuk kotak-kotak pada latar belakang grafik, sehingga memudahkan untuk mengukur jarak antara titik-titik data dan sumbu.
ax.set_ylabel('P [GPa]') #Mengatur label untuk sumbu y pada grafik menggunakan metode set_ylabel dari objek ax. Label ini menjelaskan bahwa nilai yang ditampilkan pada sumbu y adalah tekanan litostatik dalam satuan gigapascal (GPa). Notasi 'P [GPa]' memberikan informasi yang jelas tentang apa yang diukur pada sumbu y, yaitu tekanan dalam gigapascal.
ax.set_xlabel('Depth z from the Earth Surface [km]') #Mengatur label untuk sumbu x pada grafik menggunakan metode set_xlabel dari objek ax. Label ini menjelaskan bahwa nilai yang ditampilkan pada sumbu x adalah kedalaman dari permukaan bumi (z) dalam satuan kilometer (km). Notasi 'Depth z from the Earth Surface [km]' memberikan informasi yang jelas tentang apa yang diukur pada sumbu x, yaitu kedalaman dari permukaan bumi dalam kilometer.
```

```python
import numpy as np #untuk operasi matematika, terutama array dan histogram
from scipy.stats import skew #untuk menghitung skewness (kemiringan distribusi data)

a_mean = my_dataset.Zr.mean() #Menghitung rata-rata aritmatika dari kolom Zr.
median = my_dataset.Zr.median() #Menghitung median (nilai tengah) dari kolom Zr.
hist, bin_edges = np.histogram(my_dataset['Zr'], bins=20, density=True) #Membuat histogram dari data Zr dengan 20 batang, dan menghitung kepadatan probabilitasnya. hist → nilai frekuensi, bin_edges → batas-batas batang histogram
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1])/2 #Menghitung nilai modus (nilai yang paling sering muncul) dari data Zr dengan mencari batang histogram tertinggi, lalu mengambil rata-rata dari batas bawah dan batas atas batang tersebut.

standard_deviation = my_dataset['Zr'].std() #Menghitung standar deviasi dari kolom Zr, yang digunakan untuk normalisasi dalam perhitungan skewness.

a1 = (a_mean - modal_value) / standard_deviation #Menghitung Pearson's first coefficient of skewness (a1) dengan rumus: (mean - mode) / standar deviasi. Ini mengukur kemiringan distribusi data berdasarkan perbedaan antara rata-rata dan modus, dinormalisasi oleh standar deviasi.
a2 = 3 * (a_mean - median) / standard_deviation #Menghitung Pearson's second coefficient of skewness (a2) dengan rumus: 3 * (mean - median) / standar deviasi. Ini mengukur kemiringan distribusi data berdasarkan perbedaan antara rata-rata dan median, dinormalisasi oleh standar deviasi.
g1 = skew(my_dataset['Zr']) #Menghitung Fisher-Pearson's coefficient of skewness (g1) menggunakan fungsi skew dari scipy.stats, yang menghitung kemiringan distribusi data berdasarkan momen ketiga (third moment) dari data, dinormalisasi oleh standar deviasi pangkat tiga.

print ('-------') #Garis pembatas
print ("Pearson's first coefficient of skewness: {:.2f}".format(a1)) #Menampilkan Pearson's first coefficient of skewness (a1) dengan format 2 angka di belakang koma.
print ("Pearson's 2nd moment of skewness: {:.2f}".format(a2)) #Menampilkan Pearson's second coefficient of skewness (a2) dengan format 2 angka di belakang koma.
print ("Fisher-Pearson's coefficient of skewness: {:.2f}".format(g1)) #Menampilkan Fisher-Pearson's coefficient of skewness (g1) dengan format 2 angka di belakang koma.
print ('-------') #Garis pembatas lagi

'''
Output:
-------
Pearson's first coefficient of skewness: 0.74
Pearson's 2nd moment of skewness: 0.66
Fisher-Pearson's coefficient of skewness: 1.26
-------
'''
```
